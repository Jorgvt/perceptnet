```python
import re
from glob import glob

import matplotlib.pyplot as plt
import pandas as pd
```

```python
files = list(glob("./*.csv"))
files
```

```python
fig, axes = plt.subplots()
for file in files:
    name = re.findall("_\w+_(\w+)", file)
    if len(name) == 0: name = "all_free"
    else: name = name[0]
    res = pd.read_csv(file)
    axes.hist(res["pearson"], label=name, alpha=0.8)
    # break
plt.xlabel("Pearson Correlation")
plt.legend()
plt.show()
```
